# Position and Velocity

This is a presentation for Computer Science Club having to do with Kalman filters.

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


class Plot:
    def __init__(self, limits=((-10,10),(-10,10))):
        plt.xlim(limits[0][0], limits[0][1])
        plt.ylim(limits[1][0], limits[1][1])
        
    def scatter(self, x, y, color=5, area=20):
        areas = [area for i in range(len(x))]
        colors = [color for i in range(len(x))]
        plt.scatter(x, y, s=areas, c=colors, alpha=0.5)
        
    def show(self):
        plt.show() 
        

class Gaussian:
    def generate(self, params, samples=500):
        self.data = []
        for i in range(samples):
            self.data.append(np.random.normal(params[0], params[1]))
    
    def dist(self):
        return self.data
            
            
p = Plot(((-1,5),(-1000,1000)))            
            
x = Gaussian()
x.generate((1.0, 0.01))

y = Gaussian()
y.generate((0.0,300.0))
            
p.scatter(x.dist(), y.dist())    
p.show()    
    
```
