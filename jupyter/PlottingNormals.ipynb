## Plotting normal distributions

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def generate_normal(mean, std, number, result=[]):
    for i in range(number):
        result.append(np.random.normal(mean, std))
    return result
    
test = generate_normal(0.0, 1.0, 400)
test = generate_normal(4.0, 0.5, 400, test)

plt.hist(test, 30, normed=1, facecolor='green', alpha=0.75)
plt.show()
```
